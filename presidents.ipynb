```r
presidents = read.csv("presidents.csv")
```

```r
str(presidents)
```

```r
# Overall, Abe dominates, with 234 wins.
library(descr)
freq(presidents$WINNER)
```

```r
# Since 2013, when Bill came on the scene, Abe's dominance has been significantly reduced.

freq(presidents$WINNER, presidents$YEAR>=2013)
```

```r
# Since 2015, when Calvin came on the scene, Abe still leads the pack, but George has slipped behind Bill, and Tom and Calvin are cellar dwellin'. 

freq(presidents$WINNER, presidents$YEAR>=2015)
```

All told, on any given day, you'd still want to bet on Abe, but it's a much fairer fight in the last few years.
